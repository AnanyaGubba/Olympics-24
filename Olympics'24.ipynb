**Number of athletes by Discipline - BAR PLOT**

**Athletes Distribution by Country and function - TREE MAP**

Libraries used - Plotly

Description - A tree map can be used to visualize the distribution of athletes by country or function.

EDA - Bivariate

```python
import plotly.express as px

# Extracting the required top 30 counties

top_30_countries = df['country'].value_counts().nlargest(30).index
df_top_30 = df[df['country'].isin(top_30_countries)]
df_top_30 = df_top_30.rename(columns={'code': 'Athlete Count'})

# Calculate the number of athletes from each country
country_athlete_counts = df_top_30.groupby('country')['Athlete Count'].count().reset_index()
country_athlete_counts = country_athlete_counts.rename(columns={'Athlete Count': 'Athlete_Count'})

# Create the treemap
fig = px.treemap(country_athlete_counts, path=['country'],
                 values='Athlete_Count',
                 title='Top 30 Countries by Number of Athletes',
                 color='Athlete_Count',
                 color_continuous_scale='RdBu')
fig.show()
```

**Top 10 Countries by Athlete Count - STACKED BAR PLOT**

Libraries used - Matplotlib

Description - stacked bar plot, where each bar represents a country, and the segments of the bar show the distribution of male and female athletes within each country.

EDA - Multivariate

```python

top_countries = df['country'].value_counts().head(10).index
df_top_countries = df[df['country'].isin(top_countries)]
gender_country_counts = df_top_countries.groupby(['country', 'gender']).size().unstack()

plt.figure(figsize=(18, 10))
ax = gender_country_counts.plot(kind='bar', stacked=True)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Athletes", fontsize=12)
plt.title("Top 10 Countries by Athlete Count, Stacked by Gender", fontsize=14)
plt.legend(title="Gender", fontsize=10)

# Annotate bars with count values
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{int(height)}', (x + width/2, y + height/2), ha='center', va='center', fontsize=10)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend(title="Gender", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**SUNBURST**

Libraries used - Plotly

Description :

Outer Layer (Country): The largest segments represent each country in the dataset.

*   Inner Layer (Country): The largest segments represent each country in the dataset.
*   Outer Layer (Disciplines): Within each country, segments break down into disciplines (sports or types of activities).

This approach offers a quick, visually nested view of how height is distributed across different categorical levels in the dataset.

EDA - Multivariate

```python
df_top_10 = df[df['country'].isin(top_10_countries)]

athlete_counts = df_top_10.groupby(['country', 'disciplines']).size().reset_index(name='athlete_count')

df_top_10 = pd.merge(df_top_10, athlete_counts, on=['country', 'disciplines'], how='left')

# Replace zero or missing 'height' values with a small positive number
# to avoid ZeroDivisionError
df_top_10['height'] = df_top_10['height'].fillna(0.0001)
df_top_10['height'] = df_top_10['height'].replace(0, 0.0001)

fig = px.sunburst(df_top_10, path=['country', 'disciplines'], values='height', color='athlete_count',
                  title='Sunburst Chart of Top 10 Countries and Disciplines',
                  color_continuous_scale='RdBu')
fig.show()
```

**Relationship btwn the attributes age, weight and height - SCATTER MATRIX PLOT**

Libraries - Seaborn, Matplotlib, Plotly

Descripton - This Scatter matrix gives a visualization of the athletes relatesto each other in age, weight and height distingushed by their gender

EDA - Multivariate



```python

import plotly.express as px
from datetime import datetime
import plotly.graph_objects as go

df = df[df['height']>0]
df = df[df['weight']>0]

df['birth_date'] = pd.to_datetime(df['birth_date'])
df['age'] = (datetime.now() - df['birth_date']).dt.days // 365

fig = px.scatter_matrix(df, dimensions=['height', 'weight', 'age'], color='gender', title='Scatter Matrix of Height, Weight, and Age')
# fig.update_traces(diagonal_visible=False)
fig.show()
```

**Nationality, Country and Gender - SUNBURST**

Libraries used - Plotly

Description - This chart visualizes the hierarchy between nationality, country, and gender distribution.

EDA - Multivariate

```python
top_10_countries = df['country'].value_counts().nlargest(10).index
df_top_10 = df[df['country'].isin(top_10_countries)]

import plotly.express as px

# Replace NaN values in 'nationality' with 'Unknown' before creating the sunburst chart.
df_top_10['nationality'] = df_top_10['nationality'].fillna('Unknown')

# Replace NaN values in 'residence_country' with 'Unknown'
df_top_10['residence_country'] = df_top_10['residence_country'].fillna('Unknown')

# Create the sunburst chart
fig = px.sunburst(df_top_10, path=['nationality', 'residence_country', 'gender'], values='height')
fig.show()
```

**Relationship btwn Country, Discipline and Gender - PARALLEL CATEGORIES PLOT**

Libraries used - Seaborn, Matplotlib, Plotly

Description - This plot gives the count of athletes from each country participating in different disciplines separated by gender

EDA - Multivariate

```python
import plotly.express as px

top_10_countries = df['country'].value_counts().nlargest(10).index
top_10_disciplines = df['disciplines'].value_counts().nlargest(10).index
df_filtered = df[df['country'].isin(top_10_countries) & df['disciplines'].isin(top_10_disciplines)]

fig = px.parallel_categories(df_filtered, dimensions=['country', 'disciplines', 'gender'],
                             title='Categorical Relationships (Top 10 Countries, Top 10 Disciplines)')

fig.update_layout(
    font=dict(size=15),
    margin=dict(l=100, r=100, t=80, b=80),
    width=1350,
    height=600,
    )
fig.show()
```

**Athlete count of countries distinguished by gender - HEAT MAP**

Libraries used - Seaborn, Matplotlib

Description - This plot gives the frequency of athletes from each country participating in different disciplines separated by gender

EDA - Multivariate

```python
# Get the top 20 nationalities
top_20_nationalities = df['nationality'].value_counts().nlargest(20).index
df_top_20 = df[df['nationality'].isin(top_20_nationalities)]

# Create a pivot table for frequency analysis of the top 20 nationalities
pivot_table_top_20 = df_top_20.pivot_table(index='nationality', columns='gender', values='code', aggfunc='count', fill_value=0)

plt.figure(figsize=(12, 10))
sns.heatmap(pivot_table_top_20, fmt="d", cmap="magma", linewidth=0.5, annot=True)
plt.title("Frequency Heatmap of Top 20 Nationalities by Gender")
plt.xlabel("Gender")
plt.ylabel("Nationality")
plt.show()
```

**Average weight of athletes by gender for the top 10 countries - POINT PLOT**

Libraries used - Seaborn, Matplotlib

Description - This plot allows to show both the averages and variability for each country and gender.

EDA - Multivariate

```python
df = df[df['weight']>0]

top_countries = df['country'].value_counts().head(20).index
df_top_countries = df[df['country'].isin(top_countries)]

plt.figure(figsize=(12, 8))
sns.pointplot(data=df_top_countries, x='country', y='weight', hue='gender', dodge=True, ci='sd', color='#60b2ee')
plt.title("Average Weight by Gender and Country")
plt.xlabel("Country")
plt.ylabel("Average Weight (kg)")
plt.xticks(rotation=45)
plt.show()
```
